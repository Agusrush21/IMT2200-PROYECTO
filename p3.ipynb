```python
# filtrado

def filtrar(df, lect: 'str', mate: 'str'):
    filtrado = ['nom_rbd','cod_grupo','cod_rural_rbd','nom_reg_rbd','nom_com_rbd','nom_pro_rbd','cod_depe2',lect,mate]
    return df[filtrado]

# Promedio y diferencia no funciona con los datos del 2014 :(
```

```python
df14 = pd.read_excel(cuarto2014)
df14 = filtrar(df14, 'prom_lect4b_rbd', 'prom_mate4b_rbd')

# reemplazo cod_depe2
reempDepe = {
    'Municipal': '1.0',
    'Particular subvencionado': '2.0',
    'Particular pagado': '3.0',
    'Servicio Local de Educación': '4.0'
}

# reemplazo cod_grupo
reempGrupo = {
    'Bajo': '1.0',
    'Medio bajo': '2.0',
    'Medio': '3.0',
    'Medio alto': '4.0',
    'Alto': '5.0'
}

# reemplazo cod_rural
reempRural = {
    'Urbano': '1',
    'Rural': '2'
}

df14['cod_depe2'] = pd.to_numeric(df14['cod_depe2'].replace(reempDepe))
df14['cod_grupo'] = pd.to_numeric(df14['cod_grupo'].replace(reempGrupo))
df14['cod_rural_rbd'] = pd.to_numeric(df14['cod_rural_rbd'].replace(reempRural))
df14.head()
```

```python
df16 = pd.read_excel(sexto2016)
df16 = filtrar(df16, 'prom_lect6b_rbd', 'prom_mate6b_rbd')

# reemplazo cod_depe2
reempDepe = {
    'Municipal': '1.0',
    'Particular subvencionado': '2.0',
    'Particular pagado': '3.0',
    'Servicio Local de Educación': '4.0'
}

# reemplazo cod_grupo
reempGrupo = {
    'Bajo': '1.0',
    'Medio bajo': '2.0',
    'Medio': '3.0',
    'Medio alto': '4.0',
    'Alto': '5.0'
}

# reemplazo cod_rural
reempRural = {
    'Urbano': '1',
    'Rural': '2'
}

df16['cod_depe2'] = pd.to_numeric(df16['cod_depe2'].replace(reempDepe))
df16['cod_grupo'] = pd.to_numeric(df16['cod_grupo'].replace(reempGrupo))
df16['cod_rural_rbd'] = pd.to_numeric(df16['cod_rural_rbd'].replace(reempRural))
df16.head()
```

```python
df19 = pd.read_excel(octavo2019)
df19 = filtrar(df19, 'prom_lect8b_rbd', 'prom_mate8b_rbd')
df19['cod_depe2'] = df19['cod_depe2'].astype(float)
df19.head()
```

```python
df22 = pd.read_excel(segundo2022)
df22 = filtrar(df22, 'prom_lect2m_rbd', 'prom_mate2m_rbd')
df22['cod_depe2'] = df22['cod_depe2'].astype(float)
df22.head()
```

## Pregunta 3
**Existe alguna relacion entre los colegios publicos y privados, y los promedios de puntajes de SIMCE?**

Para poder responder a esta pregunta de manera efectiva, podriamos usar ya sea un scatterplot o un heatmap, los cuales podemos acceder con la libreria seaborn y matplotlib.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Los tipos de establecimientos estan bajo la columna **cod_depe2** del 1.0 al 4.0 como floats. Los numeros representan las siguientes descripciones en el glosario del ministerio de educacion:

- 1.0 = Municipal
- 2.0 = Particular Subvencionado
- 3.0 = Particular Pagado
- 4.0 = Servicio Local de Educacion

Mientras que las columnas de **prom_lect** y **prom_mate** (seguidas por su curso, ya sea 2m, 8b, 6b, etc) representan los puntajes promedios por establecimiento ese año.

## 2do Medio 2022

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df22, x='cod_depe2', y='prom_lect2m_rbd')

plt.xlabel("Codigo de Dependencia")
plt.ylabel("Promedio lectura 4to 2022")
plt.title("Comparacion de codigo de dependencia y puntaje lectura")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df22, x='cod_depe2', y='prom_mate2m_rbd')

plt.xlabel("Codigo de Dependencia")
plt.ylabel("Promedio matematica 4to 2022")
plt.title("Comparacion de codigo de dependencia y puntaje matematicas")
plt.show()
```

**Analisis**: En estos scatterplots se puede notar un incremento bastante notable en especial en **Matematicas** en los puntajes de simce mientras mas privatizado se encuentre el establecimiento.

El float 4.0 que representa el codigo de dependencia de los SLEP podria ser un dato engañoso ya que este no representa un establecimiento mas privado a pesar de tener un valor mas alto que estos.
