```python
# filtrado

def filtrar(df, lect: 'str', mate: 'str'):
    filtrado = ['nom_rbd','cod_grupo','cod_rural_rbd','nom_reg_rbd','nom_com_rbd','nom_pro_rbd','cod_depe2',lect,mate]
    return df[filtrado]

# Promedio y diferencia no funciona con los datos del 2014 :(
```

```python
df14 = pd.read_excel(cuarto2014)
df14 = filtrar(df14, 'prom_lect4b_rbd', 'prom_mate4b_rbd')

# reemplazo cod_depe2
reempDepe = {
    'Municipal': '1.0',
    'Particular subvencionado': '2.0',
    'Particular pagado': '3.0',
    'Servicio Local de Educación': '4.0'
}

# reemplazo cod_grupo
reempGrupo = {
    'Bajo': '1.0',
    'Medio bajo': '2.0',
    'Medio': '3.0',
    'Medio alto': '4.0',
    'Alto': '5.0'
}

# reemplazo cod_rural
reempRural = {
    'Urbano': '1',
    'Rural': '2'
}

df14['cod_depe2'] = pd.to_numeric(df14['cod_depe2'].replace(reempDepe))
df14['cod_grupo'] = pd.to_numeric(df14['cod_grupo'].replace(reempGrupo))
df14['cod_rural_rbd'] = pd.to_numeric(df14['cod_rural_rbd'].replace(reempRural))
df14.head()
```
