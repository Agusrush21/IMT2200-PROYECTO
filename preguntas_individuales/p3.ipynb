```python
# filtrado

def filtrar(df, lect: 'str', mate: 'str'):
    filtrado = ['nom_rbd','cod_grupo','cod_rural_rbd','nom_reg_rbd','nom_com_rbd','nom_pro_rbd','cod_depe2',lect,mate]
    return df[filtrado]

# Promedio y diferencia no funciona con los datos del 2014 :(
```

## Pregunta 3
**¿Existe una mejoría o un deterioro en los resultados del SIMCE a través de los años??**

Para poder responder a esta pregunta de manera efectiva, podriamos usar ya sea un scatterplot o un heatmap, los cuales podemos acceder con la libreria seaborn y matplotlib.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Teniendo en cuenta que hay que comparar los datos de cada año, lo mas optimo seria crear un dataframe con los promedios para simplificar la graficacion de estos.

```python
prom22 = df22[['prom_lect2m_rbd','prom_mate2m_rbd']]
prom22c = df22c[['prom_lect4b_rbd','prom_mate4b_rbd']]
prom19 = df19[['prom_lect8b_rbd','prom_mate8b_rbd']]
prom16 = df16[['prom_lect6b_rbd','prom_mate6b_rbd']]
prom16c = df16c[['prom_lect4b_rbd','prom_mate4b_rbd']]
prom14 = df14[['prom_lect4b_rbd','prom_mate4b_rbd']]

# ahora, unimos a todos

dfprom = pd.concat([prom22, prom22c, prom19, prom16, prom16c, prom14], axis=1)

dfprom.columns = [
    'Lectura_2022_2doM',
    'Matematicas_2022_2doM',
    'Lectura_2022_4toB',
    'Matematicas_2022_4toB',
    'Lectura_2019_8voB',
    'Matematicas_2019_8voB',
    'Lectura_2016_6toB',
    'Matematicas_2016_6toB',
    'Lectura_2016_4toB',
    'Matematicas_2016_4toB',
    'Lectura_2014_4toB',
    'Matematicas_2014_4toB'
]

dfprom.head()
```

Sin embargo, para hacer el analisis de estos promedios menos desordenado, es una mejor idea hacer las comparaciones por **categorias**, por ejemplo ahora empezaremos comparando los cuartos basicos de cada año.

```python
cols_4tosB = [
    'Lectura_2022_4toB',
    'Matematicas_2022_4toB',
    'Lectura_2016_4toB',
    'Matematicas_2016_4toB',
    'Lectura_2014_4toB',
    'Matematicas_2014_4toB'
]

prom4tosB = dfprom[cols_4tosB].mean()
promedios_4tosB_df = prom4tosB.reset_index()
promedios_4tosB_df.columns = ['Categoría', 'Promedio']

plt.figure(figsize=(10, 5))
sns.barplot(data=promedios_4tosB_df, x='Categoría', y='Promedio')
plt.title("Promedios de Matematicas y Lectura de 4tos basicos", fontsize=16)
plt.ylabel("Promedio", fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

Aqui se puede ver un leve incremento de promedios entre el 2014 y el 2016, **pero** los puntajes especialmente de matematicas decaen notablemente despues del periodo de pandemia o en el 2022.

```python
cols_mat = [
    'Matematicas_2022_2doM',
    'Matematicas_2022_4toB',
    'Matematicas_2019_8voB',
    'Matematicas_2016_6toB',
    'Matematicas_2016_4toB',
    'Matematicas_2014_4toB'
]

prommat = dfprom[cols_mat].mean()
prommat = prommat.reset_index()
prommat.columns = ['Categoría', 'Promedio']

plt.figure(figsize=(10, 5))
sns.barplot(data=prommat, x='Categoría', y='Promedio')
plt.title("Promedios de Matematicas desde el 2014 al 2022", fontsize=16)
plt.ylabel("Promedio", fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

Obviamente se puede apreciar el descenso de puntajes del 2019 al 2022 para cuartos basicos, aunque cabe destacar que los puntajes de 2ndo medio del 2022 se mantuvieron respecto a las generaciones pasadas. Tambien hay un salto notable entre el promedio de matematica del 6to basico del 2016 y de 8vo basico del 2019.

```python
cols_lect = [
    'Lectura_2022_2doM',
    'Lectura_2022_4toB',
    'Lectura_2019_8voB',
    'Lectura_2016_6toB',
    'Lectura_2016_4toB',
    'Lectura_2014_4toB',
]


promlect = dfprom[cols_lect].mean()
promlect= promlect.reset_index()
promlect.columns = ['Categoría', 'Promedio']

plt.figure(figsize=(10, 5))
sns.barplot(data=promlect, x='Categoría', y='Promedio')
plt.title("Promedios de Lectura desde el 2014 al 2022", fontsize=16)
plt.ylabel("Promedio", fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

Por ultimo, los cambios de los SIMCE de lectura no son tan drasticos como los de matematicas, aunque igual son notables. El peak de puntajes fue en los 4tos basicos del 2016 y a partir de ahi hubo un leve descenso hasta el 2019. Al no existir datos de los 4tos basicos de 2019 no es posible comparar la baja (o aumento) entre esos y el 4to de 2022, pero siguiendo el trend de que generalmente bajaron de puntaje despues de pandemia, no seria loco asumir que tambien existio un decremento ahi.

```python
cols = [
    'Lectura_2022_2doM',
    'Lectura_2022_4toB',
    'Lectura_2019_8voB',
    'Lectura_2016_6toB',
    'Lectura_2016_4toB',
    'Lectura_2014_4toB',
    'Matematicas_2022_2doM',
    'Matematicas_2022_4toB',    
    'Matematicas_2019_8voB',    
    'Matematicas_2016_6toB',    
    'Matematicas_2016_4toB',
    'Matematicas_2014_4toB'
]

promgen = dfprom[cols].mean()
promgen = promgen.reset_index()
promgen.columns = ['Categoría', 'Promedio']

plt.figure(figsize=(10, 5))
sns.barplot(data=promgen, x='Categoría', y='Promedio')
plt.title("Promedios de puntajes desde el 2014 al 2022", fontsize=16)
plt.ylabel("Promedio", fontsize=12)
plt.xticks(rotation=90) 
plt.show()
```

Dejando de un lado a otro los promedios de los puntajes SIMCE, se puede apreciar una forma similar en las curvas de los graficos. Esta depresion es muchisimo mas notable en los puntajes de matematicas que en los de lectura. 

Lo que pudo ser visto en todos los graficos presentados fue el brusco cambio de promedios entre el 2019 y el 2022 lo cual podria estar fuertemente vinculado a la pandemia y sus efectos en la educacion, aunque eso sera visto mas en detalle en una de las siguientes preguntas.
